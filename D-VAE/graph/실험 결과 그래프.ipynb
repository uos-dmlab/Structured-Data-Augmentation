```python
#credit card dataset, original 데이터와 100% 증강 데이터 비교
member_name =['original','CTGAN','VAE','D-VAE']
bar_width = 0.4
before_data = [99,94,96,99]
after_data = [69,94,98,99]
index = np.arange(len(member_name))

plt.bar(index, before_data, color='green',align='edge',width=bar_width,label='Accuracy')
plt.bar(index+bar_width, after_data, color='blue',align='edge',width=bar_width,label='F1-score')

plt.xticks(index+bar_width,member_name)
plt.legend()
plt.xlabel('Model')
plt.ylabel('Performance')
plt.title('CreditCard dataset(LR)')
plt.show()
```
