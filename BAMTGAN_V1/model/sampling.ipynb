##샘플링

```python
def modeling(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    #metrics(y_test,pred)
    accuracy = accuracy_score(y_test,pred)*100
    precision = precision_score(y_test,pred,average='weighted')
    recall = recall_score(y_test,pred,average='weighted')
    f1 = f1_score(y_test,pred,average='weighted')*100
    roc_score = roc_auc_score(y_test,pred,average='weighted',multi_class='ovr')*100
    print('정확도 : {0:.2f}, 정밀도 : {1:.2f}, 재현율 : {2:.2f}'.format(accuracy,precision,recall))
    print('f1-score : {0:.2f}, auc : {1:.2f}'.format(f1,roc_score,recall))
    return accuracy,roc_score,f1
    
def ml_eval(df):
  X = df.iloc[:,:-1]
  y = df.iloc[:,-1]
  X_train_clu, X_test_clu, y_train_clu, y_test_clu = train_test_split(X,y,test_size=0.25,random_state=10)

  print("SVM")
  svm = SVC(kernel='rbf', C=10)
  svm_acc,svm_auc,svm_f1 = modeling(svm,X_train_clu,X_test_clu,y_train_clu,y_test_clu)


  print("dt")
  dt = tree.DecisionTreeClassifier()
  dt_acc, dt_auc, dt_f1 = modeling(dt,X_train_clu,X_test_clu,y_train_clu,y_test_clu)

  
  print("lr")
  lr = LogisticRegression()
  lr_acc, lr_auc,lr_f1 = modeling(lr,X_train_clu,X_test_clu,y_train_clu,y_test_clu)

  print("rf")
  rf = RandomForestClassifier()
  rf_acc, rf_auc, rf_f1 = modeling(rf,X_train_clu,X_test_clu,y_train_clu,y_test_clu)

  print("knn")
  knn = KNeighborsClassifier()
  knn_acc, knn_auc, knn_f1 = modeling(knn,X_train_clu,X_test_clu,y_train_clu,y_test_clu)

  metrics_score = [[svm_acc,svm_auc,svm_f1],[dt_acc, dt_auc, dt_f1],[lr_acc, lr_auc,lr_f1],[rf_acc, rf_auc, rf_f1],[knn_acc, knn_auc, knn_f1]]
  return metrics_score
```

```python
def sampling_func(data, n_sample):
    N = len(data)
    sample = data.take(np.random.permutation(N)[:n_sample])
    return sample

def diff_eval(df_raw,df,n_sample=1,target='income'):
  df_raw_eval = ml_eval(df_raw)
  n_sample=int(len(df_raw)*2)


  for i in range(4):
    df = df.groupby(target, group_keys=False).apply(sampling_func, n_sample=n_sample)
    df = pd.concat([df,df_raw], ignore_index=True)
    print("-------------------------------------------------")
    df_eval = ml_eval(df)

    diff_results = np.array(df_raw_eval) - np.array(df_eval)
    print(diff_results)
```
