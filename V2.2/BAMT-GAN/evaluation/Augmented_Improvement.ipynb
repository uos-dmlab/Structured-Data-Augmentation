#Augmented Improvement 척도 - 최적의 증강 비율 선정

```python
import pandas as pd

def total_result(csv_path):
    # Read the data from csv file
    df = pd.read_csv(csv_path)

    # Extract augmentation ratios from the index
    augmentation_ratio = df.index / 10

    # Create a new dataframe to store the results
    result_df = pd.DataFrame()

    # Iterate over all columns in steps of 6 (as each model has 6 metrics)
    for i in range(0, df.shape[1], 6):
        model_metrics = df.iloc[:, i:i+6]

        # Calculate improvement rate compared to the baseline (first row)
        improvement_rate = model_metrics / model_metrics.iloc[0]

        # Calculate mean improvement rate and add to the results
        result_df[model_metrics.columns[0].split('_')[0] + '_improvement_rate'] = improvement_rate.mean(axis=1)

        # Copy accuracy, auc and f1 scores to the results
        result_df[model_metrics.columns[0]] = model_metrics[model_metrics.columns[0]]
        result_df[model_metrics.columns[2]] = model_metrics[model_metrics.columns[2]]
        result_df[model_metrics.columns[4]] = model_metrics[model_metrics.columns[4]]

    # Calculate 'Total Result'
    result_df['Total Result'] = result_df.mean(axis=1)

    # Scale 'Total Result' to a range of 0 to 1
    result_df['Total Result'] = (result_df['Total Result'] - result_df['Total Result'].min()) / (result_df['Total Result'].max() - result_df['Total Result'].min())

    # Set the new index
    result_df.index = ['Raw','+20%','+40%','+60%','+80%','+100%','+120%','+140%','+160%','+180%','+200%']

    return result_df['Total Result']
```

```python
total_result('../result_metrics/tox21_bamtgan.csv')
```
